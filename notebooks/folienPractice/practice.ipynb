```python
brown_universal = brown.tagged_words(tagset='universal')
print(brown_universal[:10])
print(len(set([tag for (word,tag) in brown_universal])))
```

```python
# What are the most frequent POS Tags:

brown_universal_fd = nltk.FreqDist(tag for (word,tag) in brown_universal)
print(brown_universal_fd.tabulate())
```

```python
# What are the most frequent trigrams (word)?

word_trigrams = nltk.FreqDist(nltk.trigrams(brown.words()))

print([(trigram, ':', count) for (trigram,count) in word_trigrams.most_common(10)])
```

```python
# What are the most frequent (word) trigrams that include no punctuation marks?
word_trigrams = nltk.FreqDist(nltk.trigrams(brown.words()))

for (trigram, count) in word_trigrams.most_common(40):
    if trigram[0].isalpha() and trigram[1].isalpha() and trigram[2].isalpha(): 
        print(trigram, ':', count)
        
        
```

```python
# What are the most frequent POS trigrams?

tag_trigrams = nltk.FreqDist((tok1[1], tok2[1], tok3[1]) for (tok1, tok2, tok3) in nltk.trigrams(brown_universal))
for(trigram, count) in tag_trigrams.most_common(10):
    print(trigram, ':', count)
    
    
```

```python
# What are the most frequent word trigrams with the POS sequence ADP DET NOUN?
adp_det_noun = [(tok1[0], tok2[0], tok3[0]) for (tok1, tok2, tok3) in nltk.trigrams(brown_universal) if tok1[1] == 'ADP' and tok2[1] == 'DET' and tok3[1] == 'NOUN']
for(trigram, count) in nltk.FreqDist(adp_det_noun).most_common(10):
    print(trigram, ':', count)
# array 0 = words
# array 1 = tags
```

```python
ment = [(word,tag) for (word,tag) in brown_universal if word.endswith("ment")]
print(nltk.FreqDist(tag for (word, tag) in ment).tabulate())
```

```python
# Most frequent words ending with ment that are not nouns
ment_other = [tok for tok in brown_universal if tok[0].endswith("ment") and tok[1] != "NOUN"]
ment_other_fd = nltk.FreqDist((word, tag) for (word, tag) in ment_other)
for (word, count) in ment_other_fd.most_common(10):
    print(word, ":", count)
```

```python
# Most frequent tags that are preceded by a determiner (DET)
word_tag_pairs = nltk.bigrams(brown_universal)
after_def = [tok2[1] for (tok1, tok2) in word_tag_pairs if tok1[1]=='DET']
after_def_fd = nltk.FreqDist(after_def)
print(after_def_fd.tabulate())
```
